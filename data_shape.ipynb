```python
def sid_iter(gb):
    for sid,subdf in gb:
        yield sid,subdf
```

```python
ca3_dfs = []
f0,f1,f2=True,True,True
for sid,subdf in sid_iter(gb):
    if len(subdf)==990 and f0:
        ca3_dfs.append(subdf[['Date','Id']])
        f0 = False
    if len(subdf)==942 and f1:
        ca3_dfs.append(subdf[['Date','Id']])
        f1 = False
    if len(subdf)==806 and f2:   
        ca3_dfs.append(subdf[['Date','Id']])
        f2 = False
    if not f0 and not f1 and not f2:
        print(f0,f1,f2)
        break
```

```python
len(ca3_dfs)
```

```python
ca3_plt = pd.merge(ca3_dfs[0],ca3_dfs[1],on='Date',how='outer')
ca3_plt = pd.merge(ca3_plt,ca3_dfs[2],on='Date',how='outer')
```

```python
ca3_plt.index = ca3_plt['Date']
```

```python
ca3_plt = ca3_plt.iloc[:,1:]
```

```python
ca3_plt = ca3_plt.rename(columns={'Id_x':'990','Id_y':'942','Id':'806'})
```

```python
def f(x):
    if pd.isnull(x):
        return 0
    elif x<0:
        return 2
    elif x>0:
        return 1
```

```python
ca3_plt = ca3_plt.applymap(f)
```
