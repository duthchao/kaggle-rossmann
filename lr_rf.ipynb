```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df_store = pd.read_csv('./data/store.csv',encoding='utf8')
df_train = pd.read_csv('./data/train.csv',encoding='utf8')
df_test = pd.read_csv('./data/test.csv',encoding='utf8')
```

```python
sid_feats = df_store[['Store','CompetitionDistance','Promo2']]
```

```python
dumm_feats = pd.get_dummies(df_store[['StoreType','Assortment']])
```

```python
sid_feats = pd.concat([sid_feats,dumm_feats],axis=1)
```

```python
sid_feats
```

```python
df_train = df_train[df_train.Open!=0]
```

```python
sid_ts_feats = df_train[['Store','Promo','SchoolHoliday']]
```

```python
df_train.StateHoliday = df_train.StateHoliday.map(lambda x:'0' if x==0 else x)
```

```python
df_train.DayOfWeek = df_train.DayOfWeek.map(lambda x:str(x))
```

```python
dumm_feats = pd.get_dummies(df_train[['DayOfWeek','StateHoliday']])
```

```python
sid_ts_feats = pd.concat([sid_ts_feats,dumm_feats],axis=1)
```

```python
sale_mean_df = df_train[['Store','Sales','Open']]
```

```python
sale_mean_df.index = sale_mean_df.Store
sale_mean_df = sale_mean_df[['Sales']]
```

```python
grouped = sale_mean_df.groupby(level=0)
```

```python
df = grouped.describe()
```

```python
df.loc[1].index
```

```python
store_stat_feat = pd.DataFrame(index=range(1,1116),columns=df.loc[1].index)
```

```python
for i in store_stat_feat.index:
    store_stat_feat.loc[i] = df.loc[i]['Sales']
```

```python
mer_df = pd.merge(sid_feats,sid_ts_feats,on='Store')
```
